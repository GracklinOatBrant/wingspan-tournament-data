```python
import numpy as np
import pandas as pd
import json
from collections import defaultdict
from IPython.display import HTML
```

```python
with open('wingspan_tournaments.json', 'r') as f:
    tourneys = json.load(f)
```

```python
scores_by_user = defaultdict(list)

for t in reversed(tourneys):
    for s in t['scores']:
        for user, score in s.items():
            scores_by_user[user].append(score)

score_data = pd.DataFrame([
    (k, len(v), np.median(v), np.mean(v), np.mean(v[-50:]))
    for k, v in scores_by_user.items()
 ], columns=['player', 'num_games', 'median_score', 'average_score', 'average_score_most_recent_50'])\
.set_index('player')\
.sort_values('average_score', ascending=False)
```
